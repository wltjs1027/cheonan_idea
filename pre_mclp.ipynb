```python
weight_df = pd.read_csv("final_weight.csv")
demand = pd.read_csv("demand_df.csv", encoding='utf8')
candidate = pd.read_csv("cafe_kr.csv", encoding = 'cp949')
```

```python
from geopy.distance import geodesic

def calculate_distance(temp1, temp2):
    return geodesic(temp1, temp2).kilometers

distances = []

for i in range(demand.shape[0]):
    for j in range(candidate.shape[0]):
        temp1 = (demand.loc[i, "latitude"], demand.loc[i, "longitude"])
        temp2 = (candidate.loc[j, "Latitude"], candidate.loc[j,"Longitude"])
        distance = calculate_distance(temp1, temp2)
        distances.append({
            "수요지": demand.loc[i, "이름"],
            "후보지": candidate.loc[j, "Title"],
            "거리(km)": distance,
            "행정동": demand.loc[i, "행정동"],
            "주소": demand.loc[i, "주소"]
        })
        
distances = pd.DataFrame(distances)
distances.shape
distances
```

```python
weight_df.rename(columns={"주소": "행정동"}, inplace=True)
```

```python
weight_df = weight_df[["행정동",'표준화된가중치']]
```

```python
merged_df = pd.merge(distances, weight_df, left_on='행정동', right_on='행정동', how='left')
merged_df

def standarlization(arr):
    temp_max = max(arr)
    temp_min = min(arr)
    return (arr-temp_min)/(temp_max-temp_min)+1

merged_df["표준화된거리"] = standarlization(merged_df["거리(km)"])
merged_df["표준화된거리"] = 1/merged_df["표준화된거리"]

merged_df.head()

merged_df["weighted_distance"] = merged_df["표준화된가중치"]*merged_df["표준화된거리"]
merged_df
```

```python
#가중치가 커야 중요하죠

m1 = merged_df.groupby("주소")["weighted_distance"].max()
m1 = pd.DataFrame(m1)
m1.shape
```

```python
m2 = m1.reset_index()
```

```python
from PyKakao import Local
api = Local(service_key = "alangotnocheese")

import json
import requests

def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + "nocheese"}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['x']), float(match_first['y'])

def add_lat_lon_columns(df):
    longitudes = [] #경도
    latitudes = [] #위도
    for address in df['주소']:
        try:
            lon,lat = addr_to_lat_lon(address)
            longitudes.append(lon)
            latitudes.append(lat)
        except Exception as e:
            print(f"Error processing address '{address}': {e}")
            latitudes.append(None)
            longitudes.append(None)
    df['Longitude'] = longitudes
    df['Latitude'] = latitudes
    return df
```

```python
add_lat_lon_columns(m2)
```
