***데이터 불러오기***

```python
import numpy as np
import pandas as pd
cheonan_pop = pd.read_excel('Downloads/연령별인구현황(2024.6월)-게시용 (1).xlsx', sheet_name='1. 천안시 연령별 인구(2024.6월)')
```

```python
cheonan_pop
```

```python
# 필요없는 행 (남여 구분) 삭제
cheonan_pop = cheonan_pop.dropna(axis=0)
cheonan_pop
```

```python
cheonan_pop = cheonan_pop.drop(['천 안 시','천안시 동남구','천안시 서북구'], axis=1) 
```

***각 구별 인구 합계 분포***

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#합계만 모은 변수 생성
total = cheonan_pop.iloc[0, 2:]
total
```

```python
dong_name = total.columns[:]
```

```python
total.shape
dong_name.shape
dong_name
```

```python
#인구 합계 시각화
plt.rcParams['font.family'] = 'NanumGothic'
plt.figure(figsize=(10, 6))
plt.xlabel('동 이름')
plt.ylabel('인구 수')
plt.bar(dong_name, total, color = 'red')
plt.title('천안시 인구 합계', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```

청년 인구 합친 변수 만들기
- 청년 (18~39세)
- 변수명: youth_pop

```python
cheonan_pop.columns[0]
```

```python
# '18세' 있는 인덱스 추출

row_index = cheonan_pop[cheonan_pop[cheonan_pop.columns[0]].astype(str).str.contains('18세', na=False)].index

if not row_index.empty:
    loc = cheonan_pop.index.get_loc(row_index[0])
    print(loc)
else:
    print("찾지 못함")
```

```python
cheonan_pop.iloc[[19], :]
```

찾았다!

```python
age_index = cheonan_pop.iloc[19:41, 0]
```

```python
youth_pop = cheonan_pop.iloc[19:41, 2:] #18-39세 확인
```

```python
#인덱스를 연령으로 바꾸기
youth_pop.index = [age_index]
```

```python
youth_pop
```

```python
youth_pop.shape
```

```python
total_pop = youth_pop.sum()
total_pop
```

```python
total_pop.sum()
```

**청년 인구 분포 시각화**

```python
#인구 합계 시각화
plt.rcParams['font.family'] = 'NanumGothic'
plt.figure(figsize=(10, 6))
plt.xlabel('동 이름')
plt.ylabel('인구 수')
plt.bar(dong_name, total_pop, color = 'red')
plt.title('천안시 동별 청년 인구', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()
```
