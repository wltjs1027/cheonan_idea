```python
import pandas as pd
import numpy as np

yeok = ['두정역', '성환역', '봉명역', '쌍용역', '직산역', '천안역']
latitude = [36.833705, 36.916076, 36.801215, 36.793759, 36.870593, 36.810005]
longitude = [127.14896, 127.126964, 127.135763, 127.1214, 127.143904, 127.146826]
dong = ['두정동', '성환읍', '봉명동', '쌍용1동', '직산읍', '중앙동']
cheonan_yeok = pd.DataFrame({'전철역': yeok,
                   'dong': dong,
                  'latitude': latitude, 
                  'longitude': longitude})
cheonan_yeok
```
