## 교통량에 대한 행정동별 가중치 계산하는 코드

```python
import pandas as pd

df = pd.read_csv("./data/traffic.csv", encoding="cp949")
df.head()
```

```python
df = df[['도로명', '구간명', '교통량(대)', '주소1', '주소2']]
```

```python
mylist = []

for i in df["주소1"]:
    mylist.append(i.strip())

df["주소1"] = mylist

mylist2 = []

for i in df["주소2"]:
    mylist2.append(i.strip())

df["주소2"] = mylist2
```

```python
address1 = list(pd.unique(df["주소1"]))
address1
```

```python
address2 = list(pd.unique(df["주소2"]))
address2
```

```python
adress_list = address2+address1
```

```python
final_adress_list = []
for adress in adress_list:
    final_adress_list.append(adress.strip())
```

```python
final_adress_list
```

```python
adress_list = []
for ad in final_adress_list:
    if ad not in adress_list:
        adress_list.append(ad)
```

```python
# new_df
```

```python
sub_df1 =  df[["주소1", "교통량(대)"]]
sub_df2 =  df[["주소2","교통량(대)" ]]
```

```python
sub_df1 = sub_df1.rename(columns={'주소1': '주소'})
sub_df2 = sub_df2.rename(columns={'주소2': '주소'})

sub_df = pd.concat([sub_df1, sub_df2])
```

```python
sub_df
```

```python
my_df = sub_df.groupby('주소').agg('sum')
my_df.shape
```

```python
my_df = my_df.drop(['부상동'], axis=0)
```

```python
my_df = my_df.reset_index()
my_df
```

```python
import re
new_rows = []
for _, row in my_df.iterrows():
    adress = row['주소']
    traffic = row['교통량(대)']
    
    if adress in ["원성동", "성정동", "불당동", "부성동"]:
        result = re.sub(r'동$', '', adress)
        myadress1 = f'{result}1동'
        myadress2 = f'{result}2동'
        
        traffic1 = traffic / 2
        traffic2 = traffic / 2
        
        new_rows.append({'주소': myadress1, '교통량(대)': traffic1})
        new_rows.append({'주소': myadress2, '교통량(대)': traffic2})
        
    elif adress == "쌍용동":
        result = re.sub(r'동$', '', adress)
        myadress1 = f'{result}1동'
        myadress2 = f'{result}2동'
        myadress3 = f'{result}3동'
        
        traffic1 = traffic / 3
        traffic2 = traffic / 3
        traffic3 = traffic / 3
        
        new_rows.append({'주소': myadress1, '교통량(대)': traffic1})
        new_rows.append({'주소': myadress2, '교통량(대)': traffic2})
        new_rows.append({'주소': myadress3, '교통량(대)': traffic3})
        
    else:
        new_rows.append({'주소': adress, '교통량(대)': traffic})

new_df = pd.DataFrame(new_rows)
```

```python
print(new_df)
from sklearn.preprocessing import MinMaxScaler
minmax= MinMaxScaler()

new_df['가중치'] = minmax.fit_transform(new_df[['교통량(대)']])
```

```python
new_df.sort_values("가중치", ascending=False)
```

```python
new_list = []
new_list.append({'주소': '풍세면', '교통량(대)': 0})
new_list.append({'주소': '광덕면', '교통량(대)': 0})
new_list.append({'주소': '북면', '교통량(대)': 0})
new_list.append({'주소': '성남면', '교통량(대)': 0})
new_list.append({'주소': '수신면', '교통량(대)': 0})
new_list.append({'주소': '병천면', '교통량(대)': 0})
new_list.append({'주소': '북면', '교통량(대)': 0})
new_list.append({'주소': '문성동', '교통량(대)': 0})
```

```python
new_df2 = pd.DataFrame(new_list)
```

```python
final_df = pd.concat([new_df, new_df2])
final_df
```

```python
import numpy as np

final_df = final_df.replace(np.nan, 0)
final_df
```

```python
final_df = final_df.reset_index(drop=True)
final_df
```

```python
final_df = final_df.sort_values("가중치", ascending=False)
```

```python
final_df.reset_index(drop=True, inplace=True)
```
