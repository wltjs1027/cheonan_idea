## 전체 가중치 합치는 코드

```python
import pandas as pd
import numpy as np

pm_df = pd.read_csv("./data/pm_weight.csv")
num_df = pd.read_csv("./data/cnt_df.csv")
traffic_df = pd.read_csv("./data/traffic_weight_final.csv")
```

```python
traffic_df = traffic_df.rename(columns={'가중치': 'traffic가중치'})
```

```python
merged_df = pd.merge(pm_df, traffic_df, on='주소', how='right') 
```

```python
num_df = num_df.rename(columns={'가중치': 'num가중치'})
num_df = num_df.rename(columns={'동': '주소'})
```

```python
final_df = pd.merge(merged_df, num_df, on='주소', how='left') 
```

```python
final_df = final_df[['주소','pm가중치', 'traffic가중치', 'num가중치']]
```

```python
final_df = final_df.replace(np.nan, 0)
final_df
```

```python
final_df['pm가중치'] = final_df['pm가중치']+1
final_df['traffic가중치'] = final_df['traffic가중치']+1
final_df['num가중치'] = final_df['num가중치']+1
```

```python
final_df
```

```python
final_df["가중치"] = final_df['pm가중치']*final_df['traffic가중치']*final_df['num가중치']
final_df
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax= MinMaxScaler()

final_df["표준화된가중치"] = minmax.fit_transform(final_df[['가중치']])

final_df
```

```python
final_df.to_csv("./data/final_weight.csv", index=False)
```

```python
import pandas as pd
import koreanize_matplotlib
import matplotlib.pyplot as plt

plt.bar(final_df["주소"], final_df["표준화된가중치"],color='blue')
plt.title("천안시 행정동 별 최종 가중치")  
plt.xticks(rotation=70)
plt.xlabel('행정동')
plt.ylabel('가중치')
plt.show()  
```
