```python
from geopy.distance import geodesic

def calculate_distance(temp1, temp2):
    return geodesic(temp1, temp2).kilometers

distances = []


for i in range(cheonan_cafe.shape[0]):
    for j in range(i+1, cheonan_cafe.shape[0]):
        temp1 = (cheonan_cafe.loc[i, 'Latitude'], cheonan_cafe.loc[i, 'Longitude'])
        temp2 = (cheonan_cafe.loc[j, 'Latitude'], cheonan_cafe.loc[j, 'Longitude'])
        distance = calculate_distance(temp1, temp2)
        distances.append({
            '후보지카페1': cheonan_cafe.loc[i, 'Address'],
            '후보지카페2': cheonan_cafe.loc[j, 'Address'],
            '거리(km)': distance
        })
distances = pd.DataFrame(distances)
distances.head(94)
```

```python
distances[distances["거리(km)"]==0]
distances
```

```python
distance_mean = np.mean(distances["거리(km)"])

coverage = distance_mean/2
coverage 
```

```python
distances.to_csv("/Users/soeunkim/Downloads/cheonan_distance.csv", sep=",")
```

#### 최소 커버리지 구하기 ####

```python
min_distances = []
for i in range(cheonan_cafe.shape[0]):
    for j in range(cheonan_cafe.shape[0]):
        temp1 = (cheonan_cafe.loc[i, 'Latitude'], cheonan_cafe.loc[i, 'Longitude'])
        temp2 = (cheonan_cafe.loc[j, 'Latitude'], cheonan_cafe.loc[j, 'Longitude'])
        distance = calculate_distance(temp1, temp2)
        min_distances.append({
            '후보지카페1': cheonan_cafe.loc[i, 'Address'],
            '후보지카페2': cheonan_cafe.loc[j, 'Address'],
            '거리(km)': distance
        })
        
min_distances = pd.DataFrame(min_distances)
```

```python
min_distances = min_distances[min_distances['후보지카페1'] != min_distances['후보지카페2']]
min_distances = min_distances.reset_index(drop=True)
min_distances.head(20)
```

```python
min_distances_select = min_distances.groupby('후보지카페1')['거리(km)'].min()
min_distances_select = pd.DataFrame(min_distances_select)
min_distances_select.to_csv("/Users/soeunkim/Downloads/each_min_distance.csv")
min_distances_select
```

```python
min_distance_mean = np.mean(min_distances_select["거리(km)"])
min_coverage = min_distance_mean/2
min_coverage
```
