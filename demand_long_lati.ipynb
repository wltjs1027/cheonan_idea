## 수요지에 행정동 추가
수요지: 지하철역, 행복주택, 대학교 <br>
지하철역, 행복주택, 대학교 위치에 카카오 api를 활용하여 행정동 추가하는 코드

```python
import requests
def get_address(address):
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    name = ''
    headers = {"Authorization": "KakaoAK " + ""} # 카카오 api 작성 
    try:
        name = requests.get(url, headers=headers).json()['documents'][0]['address']['region_3depth_h_name']
    except Exception as e:
        print(e)
        print(address)
    return name

def add_dong_name(df):
    dong_name = []
    for address in df['주소']:
        try:
            dong = get_address(address)
            dong_name.append(dong)
        except Exception as e:
            print(f"Error processing address '{address}': {e}")
            dong_name.append(None)
    df['행정동'] = dong_name
    return df
```

```python
import networkx as nx
import pandas as pd
import numpy as np
from PyKakao import Local
import json
import requests

def addr_to_lat_lon(addr): # 도로명 주소로 위도와 경도를 추가하는 함수
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + ""} # 카카오 api 작성
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['x']), float(match_first['y'])
    
def add_lat_lon_columns(df): # 위도와 경도 칼럼을 데이터프레임에 추가해주는 함수
    longitudes = [] #경도
    latitudes = [] #위도
    for address in df['주소']:
        try:
            lon,lat = addr_to_lat_lon(address)
            longitudes.append(lon)
            latitudes.append(lat)
        except Exception as e:
            print(f"Error processing address '{address}': {e}")
            latitudes.append(None)
            longitudes.append(None)
    df['longitude'] = longitudes
    df['latitude'] = latitudes
    return df
university_df = add_lat_lon_columns(university_df)
university_df
```
